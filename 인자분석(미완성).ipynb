### ㅇ 인자분석(요인분석)이란?  
- 다수의 변수들을 변수들 간의 관계(상관관계)를 분석하여 공통차원들을 통해 축약하는 분석기법


- 인자분석(이하 요인분석)은 관찰된 변수 집합에서 영향력있는 기본 요인 또는 잠재 변수를 
  검색하는데 사용되는 탐색적 데이터 분석 방법.  
  
  
- 변수 수를 줄여 데이터 해석에 도움을 줌.  


- 요인 분석은 논문, 시장 조사, 광고, 심리학, 금융 및 운영 연구에 널리 사용됨. 



### ㅇ 요인분석의 전제 조건
- 등간척도, 정규분포, 관찰값이 상호독립적이며 분산이 동일해야함.


- <font color=red>상관관계 행렬이 단위 행렬이라는 가설이 기각 되어야함.</font>


- 최초 요인 추출 단계에서 얻게 되는 고유치를 Scree 차트로 표현했을때, 한군데 이상에서 꺾이는 곳이 있어야 함.


- 변수들간에 높은 상관관계가 있어야함.

### ㅇ 요인분석 방법
1. 변수간의 상관행렬로부터 공통요인을 도출한다.  


2. 도출된 공통요인을 이용해서 변수간의 상관관계를 설명한다.  


3. <font color=red>요인 부하량(factor loading)은 ±0.3 이상이면 유의하다고 본다.</font>

### ㅇ 요인분석의 목적
1. 자료의 요약 : 변수들을 몇개의 공통된 변인으로 묶는다.


2. 츨정도구 <font color=red>타당성 검정</font> : 변인들이 동일한 요인으로 묶이는지를 확인.  


3. 변인구조 파악 : 변수들의 상호관계를 파악한다.  


4. <font color=red>불필요한 변인 제거 : 중요도나 설명력이 낮은 변수를 제거한다.</font> 


5. 회귀분석이나 판별분석의 설명변수 선택



### ㅇ 주성분 분석과의 차이점
- 주성분분석의 성분은 최대 분산량을 설명하는 반면 요인분석은 데이터의 <font color=red>**공분산**을 설명</font>한다.


- 주성분분석의 구성 요소는 서로 완전히 직교하는 반면 요인분석에서는 요인이 꼭 직교하는 것은 아니다.  


- 주성분분석의 성분은 *관찰된 변수의 선형 조합*이지만, 요인분석에서 관찰된 변수는 관찰되지 않은 변수 또는 요인의 선형 조합으로 모델링 된다.


### <center> $Y_{i} = \beta_{i0} + \beta_{i1} F_{1} + \beta_{i2} F_{2} + (1)e_{i}$</center>


- 주성분분석의 구성 요소는 해석 할 수 없지만, 요인분석에서 기본 요소는 라벨링 및 해석을 할 수 있다.

### ㅇ 분석을 위한 settings

```python
# 요인분석 패키지
from factor_analyzer import FactorAnalyzer
# pip3 install factor-analyzer(pip install factor-analyzer)


# 요인분석시 시각화를 위한 Tools
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd


# 요인분석을 하기위해서는 'factor-analyzer' 패키지를 설치 필요
# 요인분석 후 요인 적재량의 시각화를 위한 'seaborn'패키지를 설치 필요
```

### ㅇ 데이터 정보 불러오기

```python
#제공하는 bfi.csv파일 불러오기
df = pd.read_csv('bfi.csv', index_col=0)
df
#df.columns
#df.describe()
```

### ㅇ 데이터 정제

```python
#필요없는 열 삭제
df.drop(['gender', 'education', 'age'],axis=1,inplace=True)

#결측값(NaM)이 있는 행 삭제
df.dropna(inplace=True)

df.info()
df
```

### ㅇ 요인성 평가
- 요인분석 전 데이터셋의 요인성을 평가해야함


#### 1. Bartlett 검정
- 귀무가설: 상관관계 행렬이 단위행렬로 일정한 형태의 그래프 모양을 가지지 않는다.
- 대립가설: 상관관계 행렬이 일정한 형태의 그래프 모양을 가진다.

```python
#필요한 라이브러리 불러오기
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity

#Bartlett검정
chi_square_value,p_value = calculate_bartlett_sphericity(df)


#카이제곱, p값
chi_square_value, p_value
```

- p값이 0이므로 귀무가설을 기각하기때문에, 요인분석에 적합한 데이터라고 할 수 있음

#### 2. KMO 검정
- 변수들간의 상관관계가 다른 변수들에 의해 잘 설명되는지의 여부를 나타냄
- Bartlett과 달리 p값이 존재하지 않으며, 기준이 따로 존재함
- KMO값이 0.5 이상인 경우 설명력이 높으며 0.8이상인경우 우수하다고 할 수 있음

```python
# 필요한 라이브러리 불러오기
from factor_analyzer.factor_analyzer import calculate_kmo

# KMO 검정
kmo_all,kmo_model = calculate_kmo(df)

# KMO 값
kmo_model
```

### ㅇ 추출할 요인 수 결정하기

```python
fa = FactorAnalyzer(n_factors=25,rotation=None)
fa.fit(df)

#전체 변수의 개수는 25개, 회전은 하지 않음
#fit()는 데이터를 기반으로 선형 회귀분석 모델을 생성하는 메서드
```

```python
#생성된 모델의 Eigen값 확인
ev, v = fa.get_eigenvalues()
ev
```

```python
plt.scatter(range(1,df.shape[1]+1),ev)
plt.plot(range(1,df.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

- Eigen값이 1이상이거나 그래프 기울기가 완만해지기 전까지 나누어야 데이터 결함을 최소화 할 수 있음


- 그림과 Eigen실제값을 확인하였을때 요인은 6개로 나누어주는것이 좋음

### ㅇ 요인분석 실행

```python
fa = FactorAnalyzer(n_factors=6, rotation="varimax") 
fa.fit(df)

# 베리맥스 회전으로 돌림, ml : 최대우도 방법
```

```python
efa_result= pd.DataFrame(fa.loadings_, index=df.columns)
efa_result

#fa.loadings_ : 요인 적재량
```

- 시각화

```python
plt.figure(figsize=(6,10))
sns.heatmap(efa_result, cmap="Blues", annot=True, fmt='.2f')
```

### ㅇ 요인분석 결과
- 요인 0는 N1, N2, N3, N4 및 N5에 대해 높은 요인 적재량을 가진다.


- 요인 1는 E1, E2, E3, E4 및 E5에 대해 높은 요인 적재량을 가진다.


- 요인 2에는 C1, C2, C3, C4 및 C5에 대해 높은 요인 적재량을 가진다.


- 요인 3에는 A1, A2, A3, A4 및 A5에 대해 높은 요인 적재량을 가진다.


- 요인 4에는 O1, O2, O3, O4 및 O5 에 대해 높은 요인 적재량을 가진다.


- 요인 5에는 변수에 대해 높은 요인 적재량이 없다고 해석할 수 있다.
